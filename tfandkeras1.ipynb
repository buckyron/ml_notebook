```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
import os
import io
import pandas as pd
import requests
import numpy as np
from sklearn import metrics
```

```python

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", 
    na_values=['NA', '?'])
```

```python
cars = df['name']
```

```python
df
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
x = df[['cylinders', 'displacement', 'horsepower', 'weight',
       'acceleration', 'year', 'origin']].values
y = df['mpg'].values
```

```python
model = Sequential()
model.add(Dense(25,input_dim = x.shape[1],activation = 'relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error',optimizer = 'adam')
model.fit(x,y,verbose=2,epochs=100)
```

```python
pred = model.predict(x)
pred.shape
```

```python
print(pred)
```

```python
score = np.sqrt(metrics.mean_squared_error(pred,y))
score
```

```python
for i in range(10):
    print(f"{i+1}. Car name: {cars[i]}, MPG: {y[i]}, predicted MPG: {pred[i]}")
```

```python

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/iris.csv", 
    na_values=['NA', '?'])
```

```python
df
```

```python
x = df[['sepal_l','sepal_w','petal_l','petal_w']].values
dummies = pd.get_dummies(df['species'])
species = dummies.columns
y = dummies.values

model = Sequential()
model.add(Dense(50,input_dim = x.shape[1],activation = 'relu'))
model.add(Dense(25,activation = 'relu'))
model.add(Dense(y.shape[1],activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy',optimizer='adam')
model.fit(x,y,verbose=2,epochs=100)
```

```python
print(species)
```

```python
pred = model.predict(x)
print(f"Shape  {pred.shape}")
```

```python
print(pred)
```

```python
print(y[0:10])
```

```python
prediction = np.argmax(pred,axis=1)
expected = np.argmax(y,axis=1)
print(f"Predictions: {prediction}")
print(f"Expected: {expected}")
```

```python
print(species[prediction[1:10]])
```

```python
from sklearn.metrics import accuracy_score


correct = accuracy_score(expected,prediction)
print(f"Accuracy: {correct}")
```

```python
sample_flower = np.array( [[1.0,1.0,1.0,1.0]], dtype=float)
pred = model.predict(sample_flower)
print(pred)
pred = np.argmax(pred)
print(f"Predict that {sample_flower} is: {species[pred]}")

```

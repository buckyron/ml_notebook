```python
def smooth_mean(df1,df2,cat_name,target,weight):
    mean = df[target].mean()
    
    agg = df.groupby(cat_name)[target].agg(['count','mean'])
    counts = agg['count']
    means = agg['mean']
    
    smooth = (counts*means+weight*mean)/(counts+weight)
    
    if df2 is None:
        return df1[cat_name].map(smooth)
    else:
        return df1[cat_name].map(smooth),df2[cat_name].map(smooth.to_dict())
    
```

```python
w = 5
df['c0_enc'] = smooth_mean(df1=df,df2=None,cat_name='cat_0',target='y',weight=w)
df['c1_enc'] = smooth_mean(df1=df,df2=None,cat_name='cat_1',target='y',weight=w)
```
